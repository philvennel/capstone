Imports 


```python
from sklearn import *
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import seaborn as sns
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from pandas.plotting import scatter_matrix
%matplotlib inline
```

Importing Data


```python
# importing dataset
df= pd.read_csv('Advertising Budget and Sales.csv')
#Creating variables for use
y = df['Sales ($)'].values.reshape(-1,1)
x = df.values[:,0:3]
#Dropping the unnamed index column
df.drop(['Unnamed: 0'], axis=1, inplace=True)
myLinModel = linear_model.LinearRegression()
```

Model Training


```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size=0.15)

myLinModel.fit(x_train, y_train)

y_pred = myLinModel.predict(x_test)
y_pred #printing array of predictions

```

TV Ad budget vs Sales Visual

```python
df.plot(x='TV Ad Budget ($)', y='Sales ($)', style='o')  
plt.title('TV Ad Budget vs Sales')  
plt.xlabel('TV Ad Budget ($)')  
plt.ylabel('Sales ($)')  
plt.show()
```

Correlation between Ads and Sales Visual



```python
sns.heatmap(df.corr(), annot=True)
```

Visual for newspaper ad budget and sales with regression line

```python
sns.jointplot(x="Newspaper Ad Budget ($)", y="Sales ($)", data=df, kind="reg");
```

Simple UI for user input and prediction

```python
print("Enter the money you would like to invest in each Ad source:")
tv = float(input("TV : "))
radio = float(input("Radio : "))
newspaper = float(input("Newspaper : "))

#predicting the sales with respect to the inputs
output = myLinModel.predict([[tv,radio,newspaper]])
print("You are forecasted to get ${:.2f} (Millions) in sales."\
      .format(output[0][0] if output else "not predictable",tv,radio,newspaper))
```
